```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Assuming X and Y are your features and labels
X_train, X_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.2, random_state=42)

# Step 1: Feature Scaling
scaler = StandardScaler()

# Step 2: Combinatorial Refinement - Creating Polynomial Features
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

# Step 3: LDA for Dimensionality Reduction
lda = LDA(n_components=1) # Change n_components to 1

# Step 4: KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)

# Step 5: Create a Pipeline to Combine These Steps
pipeline = Pipeline([
    ('scaler', scaler),
    ('poly_features', poly),
    ('lda', lda),
    ('knn', knn)
])
# Step 6: Fit and Evaluate the Model
pipeline.fit(X_train, Y_train)
Y_pred_knn = pipeline.predict(X_test)

# Step 7: Evaluation Metrics
print("KNN Model Evaluation after Refinements:")
print(f'Accuracy: {accuracy_score(Y_test, Y_pred_knn):.2f}')
print(f'Precision: {precision_score(Y_test, Y_pred_knn, average="weighted"):.2f}')
print(f'Recall: {recall_score(Y_test, Y_pred_knn, average="weighted"):.2f}')
print(f'F1 Score: {f1_score(Y_test, Y_pred_knn, average="weighted"):.2f}')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Step 1: Create a Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('poly_features', PolynomialFeatures()),
    ('lda', LDA()),
    ('knn', KNeighborsClassifier())
])

# Step 2: Define Parameter Grid for Hyperparameter Tuning
param_grid = {
    'poly_features__degree': [1, 2, 3],  # Degrees of polynomial features to try
    'lda__n_components': [1],  # Since you have a binary classification, use 1 component
    'knn__n_neighbors': [3, 5, 7, 9, 11],  # Different numbers of neighbors to try
    'knn__metric': ['minkowski', 'euclidean', 'manhattan'],  # Different distance metrics
    'knn__p': [1, 2]  # p=1 is Manhattan distance, p=2 is Euclidean
}

# Step 3: Initialize GridSearchCV
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Step 4: Fit the Model
grid_search.fit(X_train, Y_train)

# Step 5: Get Best Model
best_model = grid_search.best_estimator_
print(f"Best Parameters: {grid_search.best_params_}")

# Step 6: Evaluate the Model on Test Data
Y_pred_knn = best_model.predict(X_test)

print("Best KNN Model Evaluation:")
print(f'Accuracy: {accuracy_score(Y_test, Y_pred_knn):.2f}')
print(f'Precision: {precision_score(Y_test, Y_pred_knn, average="weighted"):.2f}')
print(f'Recall: {recall_score(Y_test, Y_pred_knn, average="weighted"):.2f}')
print(f'F1 Score: {f1_score(Y_test, Y_pred_knn, average="weighted"):.2f}')
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from matplotlib.colors import ListedColormap

# Standardize the data (important before applying PCA)
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)

# Apply PCA to retain 95% of variance
pca = PCA(n_components=0.95)  # Retain 95% of the variance
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)

# Train XGBoost classifier on the reduced data
classifier = XGBClassifier()
classifier.fit(X_train_pca, Y_train)

# Predict the test set results
y_pred = classifier.predict(X_test_pca)

# Confusion matrix and accuracy
cm = confusion_matrix(Y_test, y_pred)
accuracy = accuracy_score(Y_test, y_pred)

print("Confusion Matrix:")
print(cm)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Visualize the confusion matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', cbar=False, linewidths=0.5, linecolor='black')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# Visualization of PCA components

# Visualize the decision boundary for the training set using the first two PCA components
X_set, y_set = X_train_pca[:, :2], Y_train  # Only take the first two PCA components for visualization

# Create meshgrid
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))

# Concatenate the meshgrid to be full dimensionality by adding zeros for other components
Z = np.zeros((X1.ravel().shape[0], X_train_pca.shape[1]))
Z[:, :2] = np.array([X1.ravel(), X2.ravel()]).T

# Predict on the full-dimensional meshgrid
Z_pred = classifier.predict(Z).reshape(X1.shape)

# Plot the decision boundary
plt.contourf(X1, X2, Z_pred, alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Scatter plot of training data points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('orange', 'yellow'))(i), label=j)

plt.title('XGBoost (Training set) with First 2 PCA Components')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend()
plt.show()

# Visualization for the test set

X_set, y_set = X_test_pca[:, :2], Y_test  # Only take the first two PCA components for visualization

# Create meshgrid
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))

# Concatenate the meshgrid to be full dimensionality by adding zeros for other components
Z = np.zeros((X1.ravel().shape[0], X_test_pca.shape[1]))
Z[:, :2] = np.array([X1.ravel(), X2.ravel()]).T

# Predict on the full-dimensional meshgrid
Z_pred = classifier.predict(Z).reshape(X1.shape)

# Plot the decision boundary
plt.contourf(X1, X2, Z_pred, alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Scatter plot of testing data points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('orange', 'yellow'))(i), label=j)

plt.title('XGBoost (Testing set) with First 2 PCA Components')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend()
plt.show()
```
